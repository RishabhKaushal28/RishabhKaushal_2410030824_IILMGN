```python
def add(a, b):
  return a + b
def sub(a, b):
  return a - b
def multiply(a, b):
  return a * b
def divide(a, b):
  if b == 0:
    return "invalid"
  return a / b
print("1. ADDITION")
print("2. SUBTRACTION")
print("3. MULTIPLICATION")
print("4. DIVISION")
operation = int(input("enter your choice"))
num1 = int(input("enter number"))
num2 = int(input("enter number"))
if operation == 1:
    print("Result:", add(num1, num2))
elif operation == 2:
    print("Result:", subtract(num1, num2))
elif operation == 3:
    print("Result:", multiply(num1, num2))
elif operation == 4:
    print("Result:", divide(num1, num2))
else:
    print("Invalid choice")
```

```python
def find_square_root(num):
    if num < 0:
        return "Square root not possible for negative numbers"
    return num ** 0.5
number = int(input("Enter a number: "))

result = find_square_root(number)
print("Square root:", result)
```

```python
def count_vowels(text="the life is so good"):
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    print(count)
count_vowels(text="the life is so good")
```

```python
n = int(input())
def factorial(n):
  result = 1
  for i in range(1,n+1):
    result*=i
  return result
factorial(n)
```
