```python
#array reversal
class Solution:
    def reverseArray(self, arr):
        left = 0
        right = len(arr) - 1

        while left < right:
            arr[left], arr[right] = arr[right], arr[left]
            left += 1
            right -= 1
        return arr
```

```python
class Solution:
    def kthSmallest(self, arr, k):
        arr.sort()
        kth_smallest = arr[k - 1]
        return kth_smallest
```

```python
class Solution:
    def findUnion(self, a, b):
        union_set = set()
        for num in a:
            union_set.add(num)
        for num in b:
            union_set.add(num)
        return list(union_set)
```

```python
class Solution:
    def maxSubarraySum(self, arr):
        max_sum = arr[0]
        current_sum = arr[0]
        
        for i in range(1, len(arr)):
            current_sum = max(arr[i], current_sum + arr[i])
            max_sum = max(max_sum, current_sum)
        
        return max_sum
```
